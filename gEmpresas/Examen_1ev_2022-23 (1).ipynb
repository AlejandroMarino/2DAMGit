Dibuja el siguiente sistema de ecuaciones:
$$ x+2y=3 $$
$$ x+y=2 $$

* Marca la solución con una $\star$
* Utiliza para las gráficas un ancho del linea de 1.
* Los colores son verde, rojo y azul. 
* Dibuja una linea vertical que vaya hasta la solución.
* La fuente de título es : "Arial" o "DejaVu Sans", de color darkred y tamaño 16

<img src="matplotlib.png" ALIGN="left" width=200 height=200)>


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2.5,2)
y = (3-x)/2
y2 = 2-x
font= {'family':'Arial','color':'darkred','size':16}
plt.title("Sistema de ecuaciones", loc = 'center', fontdict = font)
plt.plot(x,y,linewidth = '1',color='g', marker = 'o', linestyle=':', label=r'$y=\frac{-x+3}{2}$')
plt.plot(x,y2, linewidth = '1',color='r', marker = 'o', linestyle='--', label=r'$y={-x+2}$' )
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(x=1, ymin=0.05, ymax=0.5,color='b', marker = '*', linewidth = '1', label=r'Solución')
plt.legend()
plt.grid()
plt.show()
```

Tira tres monedas, realiza el suficiente número de tiradas para responder la siguiente pregunta.
¿Cuál sería  la probabilidad de obtener tres cruces?

```python
import random
veces = 0
tiradas = 100000
bingo = False
for i in range (tiradas): 
    t1 = random.randint(0,1)
    t2 = random.randint(0,1)
    t3 = random.randint(0,1)
    if t1 == 0 and t2 == 0 and t3 == 0:
        bingo = True
        veces += 1
print("La provavilidad es de ", veces/tiradas)
```
